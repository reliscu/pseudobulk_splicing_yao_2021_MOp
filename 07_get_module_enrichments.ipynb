Here I perform enrichment analysis to find modules enriched for cell type markers. 

These modules will later be used to correlate with exon PSI to define cell type-specific exons.

```r
mod_def <- "PosBC"
unique <- FALSE
```

## Prep DE genes

```r
pairwise_res_list <- readRDS("data/DE_genes/yao_2021_MOp_STAR_donor_cell_type_pseudobulk_pairwise_DE_genes_dream.RDS")
pairwise_ctype_genes <- prep_DE_genes(pairwise_res_list, lfc_threshold, pairwise=TRUE, unique=unique)
```

## Round 1 (pcntVar = 20, mergeParam = 0.85)

```r
network_dir <- "yao_2021_MOp_STAR_20pcntCells_20pcntVar_200samples_pseudobulk_mergeParam0.85_Modules"
dataset <- "20pcntCells_20pcntVar_200samples_pseudobulk_mergeParam0.85"
```

```r
pairwise_enrichments_df <- get_module_enrichments(network_dir, pairwise_ctype_genes, mod_def)
```

```r
# Get most enriched cell type for each module
# If cell type is most enriched in multiple modules, choose module with smallest p-value

top_mods_df <- pairwise_enrichments_df %>%
    group_by(Cell_type) %>%
    slice_min(Qval, with_ties=FALSE) %>%
    group_by(Network, Module) %>%
    arrange(Network) %>%
    slice_min(Qval, with_ties=FALSE) %>%
    arrange(Qval)
```

```r
top_mods_df[,c("Cell_type", "Pval", "Qval", "Module", "Network")]
```

```r
write.csv(top_mods_df, file=paste0("data/enrichments/yao_2021_MOp_pairwise_DE_genes_dream_unique", unique, "_", dataset, "_", mod_def, "_top_Qval_modules.csv"), row.names=FALSE, quote=FALSE)
```
