Here I make pseudobulk samples, where each sample contains a mix of cell types

```r
counts <- fread("/mnt/lareaulab/reliscu/projects/NSF_GRFP/data/scRNA-seq/yao_2021/MOp/yao_2021_MOp_STAR_gene_counts.csv", data.table=FALSE)
cell_meta <- fread("/mnt/lareaulab/reliscu/projects/NSF_GRFP/data/scRNA-seq/yao_2021/MOp/yao_2021_MOp_STAR_sampleinfo.csv", data.table=FALSE)
```

```r
table(cell_meta$subclass_label)
```

```r
all.equal(colnames(counts)[-1], cell_meta$Cell_ID)
```

```r
cell_meta$subclass_label <- sapply(cell_meta$subclass_label, function(x) gsub(" ", "_", x))
cell_meta$subclass_label <- sapply(cell_meta$subclass_label, function(x) gsub("/", "_", x, fixed=TRUE))
```

```r
# Remove cell types with < 5 cells:

min_cells <- 5

ctype_counts <- table(cell_meta$subclass_label)
low_abund_ctypes <- names(ctype_counts)[ctype_counts < min_cells]
keep <- which(!cell_meta$subclass_label %in% low_abund_ctypes)
cell_meta <- cell_meta[keep,]
counts <- counts[,c(1, keep + 1)]

all.equal(colnames(counts)[-1], cell_meta$Cell_ID)
```

```r
sort(table(cell_meta$subclass_label))
```

```r
cell.name = which(colnames(cell_meta) == "Cell_ID")
cell.type = which(colnames(cell_meta) == "subclass_label")
pcnt.cells <- 20

cell_frac <- data.frame(table(cell_meta$subclass_label) / nrow(cell_meta))
cell_meta$Freq <- cell_frac$Freq[match(cell_meta$subclass_label, as.character(cell_frac[,1]))]
cell.frac = which(colnames(cell_meta) == "Freq")
```

```r
SD <- 25

source("/mnt/lareaulab/reliscu/code/analyzeSyntheticDatasets/makeSyntheticDatasets_0.53.r")

makeSyntheticDatasets(
    expr=counts,
    sampleindex=c(2:ncol(counts)),
    cell.info=cell_meta,
    cell.name=cell.name,
    cell.type=cell.type,
    cell.frac=cell.frac,
    pcnt.cells=pcnt.cells,
    SD=SD,
    no.samples=200,
    no.datasets=1,
    with.replacement=TRUE
)
```
